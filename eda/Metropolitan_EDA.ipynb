```python
# For Metropolitan stop-and-search

import matplotlib.pyplot as plt

# Convert 'Date' column to datetime
merged_stop_and_search['Date'] = pd.to_datetime(merged_stop_and_search['Date'])

# Extract month and year from 'Date' column
merged_stop_and_search['Month_Year'] = merged_stop_and_search['Date'].dt.to_period('M')
merged_stop_and_search['Quarter'] = merged_stop_and_search['Date'].dt.to_period('Q')

# Analyze the frequency of stop and search activities over time (Monthly)
monthly_search_frequency = merged_stop_and_search['Month_Year'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
monthly_search_frequency.plot(kind='line', marker='o')
plt.title('Frequency of Stop and Search Activities Over Time (Monthly)')
plt.xlabel('Month-Year')
plt.ylabel('Number of Stop and Search Activities')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Analyze the frequency of stop and search activities over time (Quarterly)
quarterly_search_frequency = merged_stop_and_search['Quarter'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
quarterly_search_frequency.plot(kind='line', marker='o')
plt.title('Frequency of Stop and Search Activities Over Time (Quarterly)')
plt.xlabel('Quarter')
plt.ylabel('Number of Stop and Search Activities')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Investigate the demographics of individuals subjected to stop and search
gender_counts = merged_stop_and_search['Gender'].value_counts()
age_range_counts = merged_stop_and_search['Age range'].value_counts()
ethnicity_counts = merged_stop_and_search['Self-defined ethnicity'].value_counts()

# Plotting demographics
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
gender_counts.plot(kind='bar')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Number of Stop and Search Activities')

plt.subplot(1, 3, 2)
age_range_counts.plot(kind='bar')
plt.title('Age Range Distribution')
plt.xlabel('Age Range')
plt.ylabel('Number of Stop and Search Activities')

plt.subplot(1, 3, 3)
ethnicity_counts.plot(kind='bar')
plt.title('Ethnicity Distribution')
plt.xlabel('Ethnicity')
plt.ylabel('Number of Stop and Search Activities')

plt.show()

# Explore the outcomes and reasons for conducting stop and search
outcome_counts = merged_stop_and_search['Outcome'].value_counts()
legislation_counts = merged_stop_and_search['Legislation'].value_counts()

# Print unique possible outcomes and their counts
print("Unique Outcomes and Their Counts:")
print(outcome_counts)

# Print unique possible legislation and their counts
print("\nUnique Legislation and Their Counts:")
print(legislation_counts)

# Plotting outcomes and legislation
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
outcome_counts.plot(kind='bar')
plt.title('Outcome of Stop and Search')
plt.xlabel('Outcome')
plt.ylabel('Number of Stop and Search Activities')

plt.subplot(1, 2, 2)
legislation_counts.plot(kind='bar')
plt.title('Legislation for Stop and Search')
plt.xlabel('Legislation')
plt.ylabel('Number of Stop and Search Activities')

plt.tight_layout()
plt.show()
```
